```python
from sklearn.impute import SimpleImputer
data = pd.read_csv('NullDATA.csv')
i = SimpleImputer(missing_values=np.nan , strategy='most_frequent')
Char = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
i.fit(data)
Char.fit(data)
lib_data = i.transform(data)
lib_data = pd.DataFrame(lib_data, columns=data.columns)
lib_data.head()
```
